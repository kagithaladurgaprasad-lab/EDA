```python
import numpy as np
```

```python
a=np.array([[1,2,3,4],
            [2,4,6,8],
            [10,20,30,40]])
b=np.array([2,4,6,8])
print(a)
print("-"*20)
print(a.shape)
print(b.shape)
print("-"*20)
print(a+b)
print("-"*20)
print(a-b)
print("-"*20)
print(a*b)
print("-"*20)
print(a/b)
```

```python
arr1=np.arange(15).reshape(5,3)
print(arr1)
arr2=np.arange(5).reshape(5,1)
print(arr2)
print(arr1+arr2)
```

```python
a=np.array([[1,2,3,4],
            [2,4,6,8],
            [10,20,30,40]])
print(a)
b=a.T
print(b)
```

```python
a=np.array(['a','b','g'])
print(a.dtype)
```

## Rounding Functions

## 1.round()

```python
arr=np.array([3.2,3.6,7.89])
print(np.round(arr))
```

## 2.floor()

```python
arr1=np.array([3.9,2.5,-1.2,7.7])
print(np.floor(arr1))
```

## 3.ceil()

```python
arr=np.array([3.1,2.8,-1.2,4.5])
print(np.ceil(arr))
```

# where()

```python
a=np.array([40,50,70,20,85])
print(np.where(a>50,'pass','fail')) #if else logic 
print(np.where(a>50))  #it will give index positions
```

# np.any()

```python
arr=np.array([0,0,5,0])
print(np.any(arr))
arr1=np.array([0,0])
print(np.any(arr1))

a2d=np.array([[1,0],[0,0]])
print(np.any(a2d))
```
