## Random Numbers

```python
#np.random.rand()   numbers b/w 0 and 1
print(np.random.rand())

#1d array of 5 random elements
print(np.random.rand(5))

#2d array (3rows 2coloumns)
print(np.random.rand(3,2))

np.random
```

```python
#np.random.randn()
print(np.random.randn())
print("----------------")
print(np.random.randn(5))
print("----------------")
print(np.random.randn(3,2))
```

```python
# np.random.randint()
print(np.random.randint(1,10,size=5))  #1 to 9 ,no.of elements :5

#2d array
print(np.random.randint(1,10,size=(2,3)))
```

```python
#np.random.uniform(start ,stop)
print(np.random.uniform(1,10,size=5))  #1 to 9 ,no.of elements :5

#2d array
print(np.random.uniform(1,10,size=(2,3)))
```

## indexing

```python
arr=np.array([20,30,40])
print(arr)
print("----------------")
print(arr[0])
print(arr[1])
print(arr[2])
print("----------------")
print(arr[-1])
print(arr[-2])


```

```python
# slicing :extract a portion (sub_array) form the array
#arr[start:stop:stepsize)
arr=np.array([20,30,40,50,60])
print(arr[0:4])
print(arr[0:5:2])
print(arr[::-1])
```

```python
## 2d array
arr=np.array([[10,20,30],
             [40,50,60],
             [70,80,90]])
print(arr)
print('----------------')
print(arr[0,0])
print(arr[2,2])
print(arr[2,1])
print('--------------')
print(arr[-2,-2])
print(arr[-3,-3])
print(arr[-1,-1])
```

```python
arr=np.array([[10,20,30],
             [40,50,60],
             [70,80,90]])
print(arr[0,:]) #extract the first row
print(arr[1,:])
print(arr[2,:])
print('------------------')
print(arr[:,0]) #extract the colomn
print(arr[:,1])
print(arr[:,2])
print('-----------------')
print(arr[0:2,:])
print(arr[0:2,0:2])
print("----------------")
print(arr[1:3,:])
print("---------------")
print(arr[:,1:3])
```

## boolean indexing

```python
arr=np.array([1,2,3,4,5])
cond=arr>3
print(cond)
print(arr[cond])
```

```python
arr=np.array([1,2,3,4,5])
print(arr[arr>3])
```

```python
arr2d=np.array([[10,20,30],
             [40,50,60],
             [70,80,90]])
print("Condition:\n",arr2d>50)
print(arr2d[arr2d>50])
```

## Updating numpy array

```python
arr=np.array([10,20,30,40,50])
arr[0]=100
arr[1:3]=200,300
print(arr)
#fancy indexing
arr[[3,4]]=[400,500]
#update using boolean indexing
arr[arr>300]=999
print(arr)
```

```python
#updating numpy 2d array
arr2d=np.array([[10,20,30],
             [40,50,60],
             [70,80,90]])
#update singe element
arr2d[0,0]=100
print(arr2d)
print("-------------")
#update entrie row
arr2d[1]=[400,500,600]
print(arr2d)
print("------------")
#updating a portion
arr2d[0:2,0:2]=[[111,222],
                [333,444]]
print(arr2d)
print("----------")
arr2d[:,2]=[555,666,777]
print(arr2d)
print("------------")
#boolean
arr2d[arr2d>600]=999
print(arr2d)


```

```python
arr2d=np.array([[10,20,30],
             [40,50,60],
             [70,80,90]])
arr2d[-2:,-2:]=[[500,600],
                [800,900]]
print(arr2d)
print("----------")
arr2d[-3:-1,-3:-1]=[[100,200],
                    [400,500]]
print(arr2d)
```
