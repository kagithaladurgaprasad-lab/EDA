```python
#1D 2D 3D arrays

#1d(vetor)
a=np.array([1,2,3])
print("Dimension:",a.ndim)
print("shape:",a.shape) #no.of elements in tuple of integers 
print(a.size) #number of elements
print("-------------------------")
#2D (matrix)
b=np.array([[1,2,3],
           [4,5,6]])
print("Dimension:",b.ndim)
print("shape:",b.shape)
print(b.size)
print("-------------------------")
#3D--->block ,row,column
#cube or tensor ---->three dimensional array
c=np.array([[[1,2],
             [3,4]],

            [[5,6],
             [7,8]]])
print(c)
print("Dimension:",c.ndim)
print("shape:",c.shape)
print(c.size)
           
```

## NUMPY BUILT IN FUNCTIONS

```python
#np.zeros()-->creates an array filled with 0s
#np.ones()-->creates an array filled with 1s
#np.arrange()-->creates an array with a range of numbers
#np.linspace()-->creates an array with evenly spaced numbers
#np.full()-->creates an array with specific value
#np.eye()-->creates a squared identity matrix
#np.zeros()-->creates an digonal array from a list or extracts diagonal
```

```python
#np.zeros()-->creates an array filled with 0s
print("Zeros:\n",np.zeros((2,3)))

#np.ones()-->creates an array filled with 1s
print("ones:\n",np.ones((2,3)))

#np.arange()-->creates an array with a range of numbers
np.arange(1,11)

#np.linspace()-->creates an array with evenly spaced numbers
print("ones:\n",np.linspace(0,1,5))  #from 0to 1 ,it will gwnwrate 5 numbers

#np.full()-->creates an array with specific value
print("Full:\n",np.full((2,3),9))

#np.eye()-->creates a squared identity matrix
print("Eye:\n",np.eye(2))

#np.zeros()-->creates an digonal array from a list or extracts diagonal
print("dignoal:\n",np.diag([1,2,3]))

```

## dtypes

```python
arr=np.array([-1,-2,-3],dtype=np.int32)
print(arr)
print(arr.dtype)
```

```python
ar=np.array([1,2,3],dtype=np.uint32)
print(ar)
print(ar.dtype)
```

## float

```python
arr=np.array([1.5,2.5,3.5],dtype=np.float64)
print(arr)
print(arr.dtype)
```

## complex

```python
#complex64 ,complex128
arr=np.array([1+2j,2+4j],dtype=np.complex64)
print(arr)
print(arr.dtype)
```

## boolean

```python
#stores True/False
arr=np.array([True, False, True],dtype=bool)
print(arr)
print(arr.dtype)
```

## String/Unicode

```python
#stores text values
#dtype="u1"-->unicode string per element
arr=np.array(['a','b','c'],dtype="U1")
print(arr)
print(arr.dtype)
```
