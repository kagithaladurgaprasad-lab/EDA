## logical operator

```python
#logical operator
import numpy as np
a=np.array([10,20,30,40])
print(np.logical_and(a>15,a>35))
print(np.logical_or(a>15,a>35))
print(np.logical_not(a>25))
```

## assignment operator

```python
#assignment operator
a=np.array([10,20,30])
a=a+5
print(a)
a-=5
print(a)
a*=2
print(a)
a//=2
print(a)
a%=4
print(a)
a**=2
print(a)
```

## sin cos

```python
angle_deg=np.array([0,30,45,90])
#convert degrees to radians
angle_rad=np.deg2rad(angle_deg)
print("radians\n",angle_rad)
#sin values
sinval=np.sin(angle_rad)
print("sin values:",sinval)
```

```python
arr2d=np.array([[0,30],[45,90]])
angle_rad=np.deg2rad(arr2d)
print("radians\n",angle_rad)
#sin values
sinval=np.sin(angle_rad)
print("sin values:\n",sinval)
cosval=np.cos(angle_rad)
print("cos values:\n",cosval)
```

```python
angles=np.array([0,np.pi/4,np.pi/3,np.pi/2])
cos_values=np.cos(angles)
print("angles(radians):",angles)
print("cosine values:",cos_values)
```

# variance

```python
marks=np.array([50,30,70,90,60])
v=np.var(marks)
print(v)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
m=np.var(arr)
print(m)
e=np.var(arr,axis=0) #column
d=np.var(arr,axis=1)
print(e,d)
```

# standard deviation

```python
marks=np.array([50,30,70,90,60])
v=np.std(marks)
print(v)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
m=np.std(arr)
print(m)
e=np.std(arr,axis=0) #column
d=np.std(arr,axis=1)
print(e,d)
```
