```python
import numpy as np
```

# array manipulation

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.concatenate((a,b))
print(c)
```

```python
a=np.array([[1,2],[3,4]])
c=np.array([[5,6],[7,8]])
b=np.concatenate((a,c))
print(b)
```

```python
a=np.array([[1,2],[3,4]])
c=np.array([[5,6],[7,8]])
b=np.concatenate((a,c),axis=0) #row wise

d=np.concatenate((a,c),axis=1) #column wise
print(b)
print(d)
```

# vstack()


```python
a=np.array([1,2,3])
b=np.array([4,5,6])
v=np.vstack((a,b))
print(v)
```

```python
#2d 
a=np.array([[1,2],[3,4]])
c=np.array([[5,6],[7,8]])
v=np.vstack((a,c))
print(v)
```

# hstack()

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
v=np.hstack((a,b))
print(v)
```

```python
#2d 
a=np.array([[1,2],[3,4]])
c=np.array([[5,6],[7,8]])
v=np.vstack((a,c))
print(v)
```

# column stack

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
v=np.column_stack((a,b))
print(v)
```

```python
#2d 
a=np.array([[1,2],[3,4]])
c=np.array([[5,6],[7,8]])
v=np.column_stack((a,c))
print(v)
```

# split
# 1.hsplit()

```python
a=np.array([[1,2,3,8],[4,5,6,9]])

v=np.hsplit(a,2)
print(v)
```

# 2.vsplit()

```python
a=np.array([[1,2,3,8],[4,5,6,9],[11,12,13,14],[15,26,17,18]])
vs=np.vsplit(a,2)
print(vs)
```

```python
v=np.vsplit(a,4)
print(v)
```

# sorting arrays

# sort()

```python
#assending
arr=np.array([2,1,7,4])
sr=np.sort(arr)
print(arr)
print(sr)
```

```python
#descending array
print(sr[::-1])
```

```python
#2d
a=np.array([[1,2,3,8],[4,5,6,9]])
sr=np.sort(a)
print(sr)
print(sr[::-1])
```

# argsort()

```python
arr=np.array([2,1,7,4])
ar=np.argsort(arr)
print(ar)
```

```python
a=np.array([[1,2,10,8],[4,20,6,9]])
sr=np.argsort(a)
print(sr)
```
