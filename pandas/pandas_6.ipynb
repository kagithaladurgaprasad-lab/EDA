```python
#loc[row label,clo label
print(df.loc[102,"Marks"])
print(df.loc[103,'City'])
```

```python
#iloc[row index ,col index]
print(df.iloc[1,1])
print(df.iloc[0,2])
print(df.iloc[2,0])
```

```python
#at [row laberl,column_label]
print(df.at[101,'Marks'])
print(df.at[102,'City'])
```

```python
#iat[row_index,col_index]
print(df.iat[2,1])
print(df.iat[1,0])
```

## groupby(), agg()

```python
data={
    'EmpID':[1,2,3,4,5,6],
    'Name':['A','B','C','D','E','F'],
    'Department':['Sales','HR','Sales','IT','HR','IT'],
    'Salary':[50000,60000,50000,70000,60000,70000]
}
df=pd.DataFrame(data)
df
```

```python
summary=df.groupby('Department').agg(
    avr_salary=("Salary","mean"),
    toatal_salary=("Salary","sum"),
    emp_count=("EmpID","count")
)
summary
```

```python
#find max salary per department
max_salary=df.groupby("Department")["Salary"].max()
max_salary
```

```python
#find min salary per department
min_salary=df.groupby("Department")["Salary"].min()
min_salary
```

```python
mult=df.groupby("Department")["Salary"].agg(
    ['mean','sum','min','max','count','median']
)
mult
```

```python
a=df.groupby(['Department','Salary']).size()
a
```

```python
data={
    'OrderID':[1,2,3,4,5,6],
    'Region':['east','west','east','south','west','east'],
    'Product':['laptop','Moblie','laptop','tablet','laptop','mobile'],
    'Sales':[800,500,900,700,1200,400]  
}
df=pd.DataFrame(data)
byreg=df.groupby('Region')['Sales'].sum()
byreg
```

```python
avgprod=df.groupby('Product')['Sales'].mean()
avgprod
```

```python
per=df.groupby('Region')['OrderID'].count()
per
```
