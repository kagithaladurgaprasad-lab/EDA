# Series

```python
data=[10,20,30,40]
s=pd.Series(data)
print(s)
```

```python
s1=pd.Series([10,20,30],index=['a','b','c'])
print(s1)
```

```python
marks=pd.Series([67,56,89,75,95],
                index=['prasad','vinod','kaif','chandu','ram'])
print(marks)
```

```python
#access by labels
print(marks['prasad'])
```

```python
#Acess by position
print(marks[2])
```

```python
#using slicing
print(marks[0:3])
```

```python
#vectorised operation
print(marks+5)
```

```python
#aggregate functions--->sum(),min(),max(),mean(),count()
print(marks.sum())
print(marks.mean())
print(marks.min())
print(marks.max())
print(marks.count())
```

```python
#Membership
print("prasad" in marks)
print("prasad" not in marks)
```

```python
#filter
print(marks[marks>60])
```

```python
marks['prasad']=100
print(marks)
```

```python
#get index and values
print(marks.index)
print(marks.values)
```

```python
print(marks.dtype)
print(marks.shape)
print(len(marks))
print(marks.head())  # it return top 5 records /rows
print("-"*20)
print(marks.tail())
print(marks.median())
print(marks.std())
print(marks.unique())
```
