# pivot table

```python
import pandas as pd
data={
    'Region':['East','East','East','East','west','west','west','west','South','South','South','South'],
    'Product':['Laptop','Laptop','Mobile','Mobile','Laptop','Laptop','Tablet','Tablet','Moblie','Mobile','Tablet','Tablet'],
    'Sales':[800,500,600,400,1000,700,500,300,650,550,450,350],
    'Quantity':[5,3,4,2,6,4,3,2,5,3,4,2]
}
df=pd.DataFrame(data)
df
```

```python
pivot_index=pd.pivot_table(df, index ="Region",values='Sales',aggfunc='sum')
pivot_index
```

```python
pivot_column=pd.pivot_table(df,index="Region",columns='Product',values='Sales',aggfunc='sum')
pivot_column
```

```python
pivot_fill=pd.pivot_table(df,index="Region",columns='Product',values='Sales',aggfunc='sum',fill_value=0)

pivot_fill
```

```python
pivot_fill=pd.pivot_table(df,index="Region",columns='Product',values='Sales',aggfunc=['min','max'],fill_value=0)

pivot_fill
```

```python
pivot_column=pd.pivot_table(df,index="Region",columns='Product',values='Sales',aggfunc='count',fill_value=0)
pivot_column
```

```python
pivot_sum=pd.pivot_table(df,index="Region",columns='Product',values='Sales',aggfunc='sum',fill_value=0,margins=True)
pivot_sum
```

## Crosstab

```python
data ={
    'Employee':['jhon','james','david','mary','tom','charlie'],
    'Department':['Sales','HR','Sales','IT','HR','IT'],
    'Gender':['Male','Female','Male','Female','Male','Male'],
    'Salary':[50000,60000,50000,70000,62000,68000]
}
df=pd.DataFrame(data)
df
```

```python
#count of employes per department x genderc
ct=pd.crosstab(index=df['Department'],columns=df['Gender'])
ct
```

```python
ct_sum=pd.crosstab(index=df['Department'],columns=df['Gender'],values=df['Salary'],aggfunc='sum')
ct_sum
```

```python
ct_mult=pd.crosstab(index=df['Department'],columns=df['Gender'],values=df['Salary'],aggfunc=['sum','mean'])
ct_mult
```

```python
ct_mult=pd.crosstab(index=df['Department'],columns=df['Gender'],values=df['Salary'],aggfunc=['sum','mean'],margins=True)
ct_mult
```

```python
ct_norm=pd.crosstab(index=df['Department'],columns=df['Gender'],normalize='index')
ct_norm
```

## CONCATENATION

```python
data_a={
    'EmpID':[1,2],
    'Name':['Jhon','Tom'],
    'Department':['Sales','HR']
}
data_b={
    "EmpID":[3,4],
    "Name":["David","Tom"],
    "Department":["IT","Sales"]
}
df_a=pd.DataFrame(data_a)
df_b=pd.DataFrame(data_b)
print(df_a)
print('-----------------')
print(df_b)      
```

```python
#row wise concatenation
row_con=pd.concat([df_a,df_b],axis=0,ignore_index=True)
row_con
```

```python
#column wise concatenation
data_a={
    'EmpID':[1,2],
    'Name':['Jhon','Tom'],
}
data_b={
    "Department":["IT","Sales"],
    'salary':[50000,60000]
}
df_a=pd.DataFrame(data_a)
df_b=pd.DataFrame(data_b)
print(df_a)
print('-----------------')
print(df_b)      
```

```python
col_concat=pd.concat([df_a,df_b],axis=1)
col_concat
```

# merging/joins

```python
employees=pd.DataFrame({
    'EmpID':[1,2,3,4],
    'EmpName':['A','B','c','D'],
    'DeptID':[101,102,103,104]
})
departments=pd.DataFrame({
    'DeptID':[101,102,103,105],
    'DeptName':['HR','IT','Finance','Marketing']
})
print(employees)
print('-------------------------')
print(departments)
```

```python
# Inner join ----> it will returns common rows only
inner_join=pd.merge(employees,departments,on='DeptID',how='inner')
print(inner_join)
```

```python
# left join ----> all employees + matching depts
left_join=pd.merge(employees,departments,on='DeptID',how='left')
print(left_join)
```

```python
# Right join ----> all departments + matching emps
right_join=pd.merge(employees,departments,on='DeptID',how='right')
print(right_join)
```

```python
# outer join ----> all records from df's
outer_join=pd.merge(employees,departments,on='DeptID',how='outer')
print(outer_join)
```
