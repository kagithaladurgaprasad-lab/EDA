# Selecting and swapping columns

```python
df=pd.DataFrame({
    'Name':['a','b','c'],
    'Age':[23,25,21],
    'City':['Delhi','Mumbai','Chennai']
})
df
```

```python
print(df["Name"])
print('----------------')
print(df[['Name','City']])
print('------------------')
print(df.iloc[:,0]) #first column
print('------------------')
print(df.iloc[:,[0,2]]) #first and third column
print('------------------')
print(df.iloc[:,0:2])  # first two columns
      
```

```python
#swapping
#manual order
df1=df[['Name','City','Age']]
print(df1)
print('---------------')
#swap two columns(2nd method)
cols=list(df.columns)
cols[1],cols[2]=cols[2],cols[1]
df2=df[cols]
print(df2)
print('---------------')

#3rd method---reindex
df3=df.reindex(columns=['City','Name','Age'])
print(df3)
print('---------------')

#4th method
df4=df.iloc[:,[2,0,1]]
print(df4)
```

# adding new column and drop exisiting column

```python
df=pd.DataFrame({
    'Name':['a','b','c'],
    'Age':[23,25,21],
    'City':['Delhi','Mumbai','Chennai']
})
df
```

```python
#add new column
df['country']='India'
df
```

```python
df['marks']=[85,90,94]
df
```

```python
df["Age_after_5years"]=df["Age"]+5
df
```

```python
df=df.assign(Status=['pass','pass','pass'])
df
```

```python
df=df.drop("marks",axis=1)
df
```

```python
df2=df.drop(['country','Status'],axis=1)
df2
```

```python
df.drop('Status',axis=1,inplace=True) #inplace =modifies the original data frame directly ,no need to reassign
df  
```

# Vectorized Arthematic Operations in pandas

```python
df=pd.DataFrame({
    'A':[10,20,30],
    'B':[1,2,3]
})
df
```

```python
df['A'+'B']=df['A']+df['B']
df
```

```python
df['A-B']=df['A']-df['B']
df
```

```python
df['A*B']=df['A']*df['B']
df
```

```python
df['A/B']=df['A']/df['B']
df
```

```python
df['A%B']=df['A']%df['B']
df
```

```python
df['A**B']=df['A']**df['B']
df
```

```python
df['A+5']=df['A']+5
df['B*10']=df["B"]*10
df
```
