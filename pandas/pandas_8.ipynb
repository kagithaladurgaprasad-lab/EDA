# index based merging

```python
import pandas as pd
```

```python
employees=pd.DataFrame({
    'Empname':['A','B','C','D'],
    'Dept':['HR','IT','Finance','Sales']
},index=[101,102,103,104])

salaries=pd.DataFrame({
    'Salary':[50000,60000,55000,45000]
},index=[101,102,103,105])
print(employees)
print('-------------------------')
print(salaries)
```

```python
#inner_join
inner=employees.merge(salaries,left_index=True,right_index=True,how='inner')
inner
```

```python
#left join
left=employees.merge(salaries,left_index=True,right_index=True,how='left')
left
```

```python
#Right join
right=employees.merge(salaries,left_index=True,right_index=True,how='right')
right
```

```python
#outer join
outer=employees.merge(salaries,left_index=True,right_index=True,how='outer')
outer
```

# Data Cleaning/Handling Missing Data

```python
data={
    'Customer':['A','B','C','D','E'],
    'Age':[25,None,30,None,28],
    'City':['New York','Los Angles',None,'Chicago',None],
    'Purchase':[200,None,150,300,None]
}

df=pd.DataFrame(data)
df
```

```python
#Detect missimg values
#true--missing values
#false--non missing values
df.isnull()
```

```python
#count of total missing values in each column
df.isnull().sum()
```

```python
#count of total missing values
df.isnull().sum().sum()
```

```python
#Dectct non misssing values
#false--missing values
#true--non missing values
df.notnull()
```

```python
#count of non missing values in each column
df.notnull().sum()
```

```python
#total non missing values in dataset
df.notnull().sum().sum()
```

```python
df1=df.fillna(0)
df1
```

```python
df2=df.fillna({'Age':df['Age'].mean(),'City':'unknown','Purchase':df['Purchase'].median()})
df2
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df_ffill=df.fillna(method='ffill')
df_ffill
```

```python
df_bfill=df.fillna(method='bfill')
df_bfill
```

```python
#dropna()
#by default it will delete records (axis=0)
df3=df.dropna()
df3
```

```python
# for columns
df4=df.dropna(axis=1)
df4
```

```python
#threshold--->keeps rows with atleast 3 non missing values
df5=df.dropna(thresh=3)
df5
```

```python
#delate row with tleast one NaN
df6=df.dropna(subset=['Age','Purchase'])
df6
```
