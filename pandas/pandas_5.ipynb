# Data Type Conversion








#astype()---convert one data type to another 



```python
import pandas as pd
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.DataFrame({
    "Age":['23','24','25'],
    "salary":[25000.6,30000.4,50000.8],
    "Employed":[1,0,1]
})
```

```python
df.info()
```

```python
#convert age from object to int
df['Age']=df['Age'].astype(int)
df.dtypes
```

```python
df['salary']=df['salary'].astype(int)
df.dtypes
```

```python
df['Employed']=df['Employed'].astype(bool)
df.dtypes
```

```python
# pd.to_numeric
df['Age']=df['Age'].astype(str)
df.dtypes
df['Age']=pd.to_numeric(df["Age"],errors="coerce")
print(df.dtypes)
```

```python
df['joinDate']=pd.to_datetime(['2025-09-09','2024-01-01','2023-12-15'])
df.dtypes
```

```python
df['experiance']=pd.to_timedelta(["365 days","730 days","1095 days"])
df.dtypes
```

```python
df["department"]=pd.Series(["HR","IT","Finance"]).astype("category")
df.dtypes
```

# Boolean Row Selection by Values

```python
data={
    "Name":['A','B','C','D','E','F'],
    "Age":[25,30,35,40,45,50],
    "City":["New york","Paris","London","Berlin","paris","London"],
    "Salary":[50000,60000,70000,80000,90000,100000]
}
df=pd.DataFrame(data)
df
```

```python
df[df["Age"]==30]
```

```python
df[df["Age"]!=30]
```

```python
# AND Condition (&)
df[(df['Age']>30)& (df['City']=='paris')]
```

```python
#OR condition(|)
df[(df['Age']<30)|(df['City']=='paris')]
```

```python
# isin() condition
df[df["City"].isin(["Paris","Hyderabad"])]
```

```python
#not in --->(~with isin())
df[~df["City"].isin(["Paris","Hyderabad"])]  #give other than  paris and hydrabad
```

```python
 #contains
df[df['City'].str.contains("on")]
```

```python
 #contains
df[~df['City'].str.contains("on")]
```

```python
#Between contition
df[df["Age"].between(30,45)]
```

```python
# not Between contition
df[~df["Age"].between(30,45)]
```

```python
#startswith 
df[df['City'].str.startswith("N")]
```

```python
#endswith
df[df['City'].str.endswith("n")]
```

```python
#age>30,city=london,salary>=70000
df[(df['Age']>30)& (df['City']=='paris')&(df['Salary']>=70000)]
```
