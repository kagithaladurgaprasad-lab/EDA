## DESCRIPTIVE STATISTICS

```python
import statistics as st
marks=[50,60,60,70,80]
mean_value=st.mean(marks)
meadian_value=st.median(marks)
mode_value=st.mode(marks)

print("Mean: ",mean_value)
print("Meadian: ",meadian_value)
print("Mode: ",mode_value)
```

## MEASURE OF DISPERSION

```python
import statistics as st
data =[48,49,50,51,52]

#range
data_range=max(data)-min(data)
print("Range:" ,data_range)

#population varation and standeard deviation
pop_variance=st.pvariance(data)
pop_std=st.pstdev(data)

print("Population Variance:",pop_variance)
print("Population standard deviation:", pop_std)

#sample varicance and standaard devation
sample_variance=st.variance(data)
sample_std=st.stdev(data)

print("sample Variance:",sample_variance)
print("sample standard deviation:", sample_std)

```

## quartiles

```python
data=[5,10,15,20,25,30,35,40,45,50,55,60]
data.sort()
#median
def median(lst):
    n=len(lst)
    mid=n//2
    if n %2==0:
        return (lst[mid-1]+lst[mid])/2
    else:
        return lst[mid]

Q2=median(data)

lower=data[:len(data)//2] #Q1
upper=data[len(data)//2:] #Q3

Q1=median(lower)
Q3=median(upper)

IQR=Q3-Q1

Q0=min(data)
Q4=max(data)

print("Q1:",Q1)
print("Q2:",Q2)
print("Q3:",Q3)
print("IQR:",IQR)
print("Q0:",Q0)
print("Q4:",Q4)


```

## CORRELATION

```python
import statistics as st
study_hours=[2,4,6,8,10]
exam_scores=[40,50,60,70,80]

#mean
mean_x=st.mean(study_hours)
mean_y=st.mean(exam_scores)

#calculator numerator

numerator=sum((x-mean_x)*(y-mean_y) for x,y in zip(study_hours,exam_scores))

#calculate denominator

denominator=(sum((x-mean_x)**2 for x in study_hours)* sum((y-mean_y)**2 for y in exam_scores))**0.5

#calculate correlation
r=numerator/denominator
print ("Correlation(r)= ",round(r,2))
```

## RANK CORRELATION

```python
math_rank=[1,2,3,4,5]
science_rank=[2,1,4,3,5]

d_square=[(x-y)**2 for x,y in zip(math_rank,science_rank)]
sum_d_square=sum(d_square)
n=len(math_rank)

r_s=1-(6*sum_d_square)/(n*(n**2-1))

print("Rank correlation: ",round(r_s,2))
```
